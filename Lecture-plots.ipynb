```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Exercise 0: Pattern matching

```python
x = np.linspace(-np.pi,np.pi)
```

```python
for i,(y,title) in enumerate([(np.sin(x),'$\sin(x)$'),(np.cos(x),'$\cos(x)$'),(np.ones_like(x),'$1$'),(x,'$x$'),(x**2,'$x^2$')]):
    plt.plot(x,y,label=title)
    
    plt.text(0,4,title,ha='center',fontsize=36)
    plt.xlabel('$x$',fontsize=20)
    plt.ylabel('$y$',fontsize=20)
    plt.xlim(-3,3)
    plt.ylim(-3,6)
    plt.savefig(f'figures/fig{i}.jpg')
    plt.show()
```

```python
N = 100

xi = np.random.uniform(-np.pi,np.pi,N)
eps = np.random.randn(N,)
yi =xi**2 + 0.5*eps
```
